```python
churn = data[data['Churn'] == 'No']
no_churn = data[data['Churn'] == 'Yes']
```

```python
def churn_nochurn (column_name):
    fig,axs = plt.subplots(2, figsize = (10,10))
    axs[0].pie(
        x= list(churn[column_name].value_counts().values),
        labels = list(churn[column_name].value_counts().keys()),
        autopct='%1.1f%%')
    axs [0].set_title('Churn', fontweight = 'bold')
    
    axs[1].pie(
        x= list(no_churn[column_name].value_counts().values),
        labels = list(no_churn[column_name].value_counts().keys()),
        autopct='%1.1f%%')
    axs [1].set_title('Churn', fontweight = 'bold')
```

```python
churn_nochurn('gender')
```

# PRE-PROCESSING

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
scaled_num = pd.DataFrame (ss.fit_transform(df_num), columns = num)
```

```python
scaled_num.describe()
```

```python
encoder = OneHotEncoder()
encoded_cat = pd.DataFrame(encoder.fit_transform(df_cat).toarray(), columns = encoder.get_feature_names())
encoded_cat.drop(columns = {
    'x16_No', 'x16_Yes'}, inplace = True)
encoded_cat.head()
```

```python
cleaned = encoded_cat.merge(scaled_num,left_index=True,right_index=True,how = "left")
cleaned['Churn'] = data['Churn']
cleaned['Churn'].replace({'Yes' : 1, 'No' : 0}, inplace = True)
cleaned.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(
    cleaned.drop('Churn', axis =1 ), cleaned['Churn'], test_size=0.3, random_state = 5)

print (X_train.shape)
print (X_test.shape)
print (Y_train.shape)
print (Y_test.shape)
```

# Decision Tree

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=2)
model.fit(X_train, Y_train)

dt_pred_train = model.predict(X_train)
dt_pred_test = model.predict(X_test)
```

```python
dt_pred = pd.DataFrame(zip(dt_pred_train, Y_train), columns = ['Prediction', 'Actual'])
dt_pred
```

# Decision Tree - Evaluation

```python
tree.plot_tree(model,
               filled=True,
               feature_names = list(X_train.columns), 
               class_names=['Churn', 'No Churn'],)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn import metrics
print ("\n Classification report : \n",classification_report(Y_train,dt_pred_train))
print ("Accuracy   Score : ",accuracy_score(Y_train,dt_pred_train))
```

```python
print ("\n Classification report : \n",classification_report(Y_test,dt_pred_test))
print ("Accuracy   Score : ",accuracy_score(Y_test,dt_pred_test))
```

```python
plt = sns.heatmap(confusion_matrix(Y_test, dt_pred_test), annot=True)
plt.set_title('Confusion Matrix of Decision Tree Model')
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=20, random_state=1, max_depth=3)
rf.fit(X_train, Y_train)
rf_pred_train = rf.predict(X_train)
rf_pred_test = rf.predict(X_test)
```

```python
rf_pred = pd.DataFrame(zip(rf_pred_train, Y_train), columns = ['Prediction', 'Actual'])
rf_pred
```

# Random Forest - Evaluation

```python
print ("\n Classification report : \n",classification_report(Y_train, rf_pred_train))
print ("Accuracy   Score : ",accuracy_score(Y_train, rf_pred_train))
```

```python
print ("\n Classification report : \n",classification_report(Y_test, rf_pred_test))
print ("Accuracy   Score : ",accuracy_score(Y_test, rf_pred_test))
```

```python
plt = sns.heatmap(confusion_matrix(Y_test, rf_pred_test), annot=True)
plt.set_title('Confusion Matrix of Random Forest Model')
```
