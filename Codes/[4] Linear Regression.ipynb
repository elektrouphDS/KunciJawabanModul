# Importing Libraries + Reading Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn

from sklearn.datasets import load_boston
boston = load_boston()
print(boston.DESCR)
```

```python
data = pd.DataFrame(boston.data)
data.columns = boston.feature_names
data.head()
```

```python
data['PRICE'] = boston.target
data.head()
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(
    data.drop('PRICE', axis =1 ), data['PRICE'], test_size=0.3, random_state = 5)

print (X_train.shape)
print (X_test.shape)
print (Y_train.shape)
print (Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```

```python
coef = pd.DataFrame(zip(X_train.columns, model.coef_), columns = ['Features', 'Coefficients'])
coef
```

```python
X_test
```

```python
diff = pd.DataFrame(zip(pred_test, Y_test), columns = ['Predicted Prize', 'Real Prize'])
diff
```

# Evaluation

```python
plt.figure(figsize=(10,10))
sns.regplot(diff['Predicted Prize'], diff['Real Prize'])
```

### Dump to pickle for Materi 6

## Sum Residual & Mean Square Error (MSE)

```python
residu = np.sum((Y_test - pred_test) ** 2)
mse = np.mean((Y_test - pred_test) ** 2)
print(mse, residu)
```

# New Analysis with Data Cleansing 

```python
data
```

```python
#CHAS tidak di buang karena binary variable
chas = data['CHAS']
data = data.drop('CHAS', axis =1)
```

```python
data.plot(kind = 'box', subplots= True, layout = (5,5))
plt.show()
```

```python
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)
iqr = q3-q1
print(iqr)
```

```python
condition_1 = (q1 - 1.5 * iqr)
condition_2 = (q3 + 1.5 * iqr)

data_cleaned = data[~((data < condition_1) | (data > condition_2)).any(axis=1)]
data_cleaned['CHAS'] = chas
```

```python
sns.set(font_scale= 1.5, rc={'figure.figsize':(25,25)})
ax = data_cleaned.hist(bins=25)
```

```python
fig, axes = plt.subplots (5,3)

# Crime Rate vs PRICE
axes [0,0].scatter (data_cleaned['CRIM'], data_cleaned ['PRICE'])
axes [0,0].set_title('Crime Rate vs PRICE', fontweight = 'bold')

# ZN vs PRICE
axes [0,1].scatter (data_cleaned['ZN'], data_cleaned ['PRICE'])
axes [0,1].set_title('ZN vs PRICE', fontweight = 'bold')

# INDUS vs PRICE
axes [0,2].scatter (data_cleaned['INDUS'], data_cleaned ['PRICE'])
axes [0,2].set_title('INDUS vs PRICE', fontweight = 'bold')

# CHAS vs PRICE
axes [1,0].scatter (data_cleaned['CHAS'], data_cleaned ['PRICE'])
axes [1,0].set_title('CHAS vs PRICE', fontweight = 'bold')

# NOX vs PRICE
axes [1,1].scatter (data_cleaned['NOX'], data_cleaned ['PRICE'])
axes [1,1].set_title('NOX vs PRICE', fontweight = 'bold')

# RM vs PRICE
axes [1,2].scatter (data_cleaned['RM'], data_cleaned ['PRICE'])
axes [1,2].set_title('RM vs PRICE', fontweight = 'bold')

# AGE vs PRICE
axes [2,0].scatter (data_cleaned['AGE'], data_cleaned ['PRICE'])
axes [2,0].set_title('AGE vs PRICE', fontweight = 'bold')

# DIS vs PRICE
axes [2,1].scatter (data_cleaned['DIS'], data_cleaned ['PRICE'])
axes [2,1].set_title('DIS vs PRICE', fontweight = 'bold')

# RAD vs PRICE
axes [2,2].scatter (data_cleaned['RAD'], data_cleaned ['PRICE'])
axes [2,2].set_title('RAD vs PRICE', fontweight = 'bold')

# TAX vs PRICE
axes [3,0].scatter (data_cleaned['TAX'], data_cleaned ['PRICE'])
axes [3,0].set_title('TAX vs PRICE', fontweight = 'bold')

# PTRATIO vs PRICE
axes [3,1].scatter (data_cleaned['PTRATIO'], data_cleaned ['PRICE'])
axes [3,1].set_title('PTRATIO vs PRICE', fontweight = 'bold')

# B vs PRICE
axes [3,2].scatter (data_cleaned['B'], data_cleaned ['PRICE'])
axes [3,2].set_title('B vs PRICE', fontweight = 'bold')

# LSTAT vs PRICE
axes [4,0].scatter (data_cleaned['LSTAT'], data_cleaned ['PRICE'])
axes [4,0].set_title('LSTAT vs PRICE', fontweight = 'bold')

plt.show()
```
